```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 998.8
sigma = 73.10
x1 = 900
x2 = 1100

z1 = ( x1 - mu ) / sigma
z2 = ( x2 - mu ) / sigma

x = np.arange(z1, z2, 0.001) 
x_all = np.arange(-10, 10, 0.001)
```

```python
y = norm.pdf(x,0,1)
y2 = norm.pdf(x_all,0,1)

fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('seaborn-whitegrid')
ax.plot(x_all,y2)

ax.set_xlim([-4,4])
ax.set_yticklabels([])

plt.text(-3.57, 0.425, r'Mesokurtic Distribution', {'color': 'black', 'fontsize': 12})
plt.text(-.57, 0.425, r'$\mu = 0, \sigma^2 = 0$', {'color': 'black', 'fontsize': 12})

plt.show()
```

```python
y = norm.pdf(x,0,0.5)
y2 = norm.pdf(x_all,0,0.5)

fig, ax = plt.subplots(figsize=(9,6))
plt.style.use('seaborn-whitegrid')
ax.plot(x_all,y2)

ax.set_xlim([-4,4])
ax.set_yticklabels([])

plt.text(-3.57, 0.85, r'Leptokurtic Distribution', {'color': 'black', 'fontsize': 12})
plt.text(-.57, 0.85, r'$\mu = 0, \sigma^2 = 0.5$', {'color': 'black', 'fontsize': 12})

plt.show()
```
